```python
import numpy as np
import pandas as pd
from pymatreader import read_mat
import os.path
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
df = read_mat('T215_4_000_1.mat')
df.keys()
```

```python
data = df['Uh_AverageWindSpeed']
data
location = pd.DataFrame(df['Location_of_measured_points'])
location
```

```python
df['Readme']
```

```python
location.T
```

```python
cp  = pd.DataFrame(df['Wind_pressure_coefficients'])
meancp = cp.mean()
location_t = location.T
dat = pd.concat([location_t,meancp],axis=1,ignore_index=True)

angle = pd.Series([0]*(len(dat)))

meandat =pd.concat([dat,angle],axis=1,ignore_index=True)
meandat
```

```python
faces = meandat.groupby(meandat[3])
dfs = {}
for name, group in faces:
    dfs[name] = group
face1 = dfs[1]
face1.reset_index(drop=True,inplace=True)
face1
```

```python
traindat = face1[[0,1,5,4]]
traindat.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

X = traindat[[0,1,5]].copy()
y = traindat[4]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1, random_state=42)
model = SVR(kernel='rbf', C=100, gamma='scale')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R-squared: {r2:.2f}')
print(f'MSE: {mse:.2f}')
X
```

```python
cont = face1[[0,1,4]].copy()
cont.rename(columns={0: "x", 1: "y",4:"cpm"},inplace=True)
Z = cont.pivot_table(index='x', columns='y', values='cpm').T.values
X_unique = np.sort(cont.x.unique())
Y_unique = np.sort(cont.y.unique())
pd.DataFrame(Z)
```

```python
import matplotlib.pyplot as plt
plt.contourf(X_unique,Y_unique,Z,[.1,.2,.3,.4,.5,.6,.7,.8,.9],cmap='autumn')
plt.colorbar()
```
